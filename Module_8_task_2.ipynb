```python
import pandas as pd
import numpy as np

data = pd. read_csv('C:/users/pfali/Downloads/fatal-police-shootings-data.csv')
data['race'].fillna(value="data missing",inplace=True)
```

```python
df=data.pivot_table(values='signs_of_mental_illness',index='race', aggfunc=np.sum)
print(df)
```

```python
races=data['race'].unique().tolist()
list=[]
for race in races:
    x=len(data[data['race']==race])
    list.append(x)

race_mental=dict(zip(data['race'].unique(),list))

df['race']= ['A','B','H','N','O','W','data missing']

df['per_cent_mental_illness']=df['race'].map(race_mental)
df['per_cent']=df['signs_of_mental_illness']/df['per_cent_mental_illness']
print(df)
print("Najwyższy odsetek znamion chorób psychicznych charakteryzuje rasę W.")
```

```python
data['date']=pd.to_datetime(data['date'])

arr=np.ones((5416,1))
data['no_of_crimes']=pd.DataFrame(arr)

data['day_name']=data['date'].dt.day_name()

data['day_name'].replace(to_replace=["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday"],value=["poniedziałek","wtorek","środa","czwartek","piątek","sobota","niedziela"],inplace=True)

crimes_day=data.pivot_table(values='no_of_crimes',index='day_name',aggfunc=np.sum)
crimes_day['day_name']=['czwartek','niedziela','piątek','poniedziałek','sobota','wtorek','środa']

order=['poniedziałek','wtorek','środa','czwartek','piątek','sobota','niedziela']
data.pivot_table(values='no_of_crimes',index='day_name',aggfunc=np.sum).loc[order].plot(kind='bar')
```
