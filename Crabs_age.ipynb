Observations:
- we have one numeric value for 'sex' - to be replaced with mode of this feature
- females and males crabs don't occur in the groups of the youngest age; youngest crabs are of indeterminate sex, but at the same time they aren't the oldest ones
- we have two distant observations of length, which are outliers and should be deleted
- values over 2.0 of diameter look like outliers
- values over 1.0 of height look like outliers
- values over 20.0 of viscera weight look like outliers
- values over 35.0 of shell weight look like outliers
- height, diameter and length of values 0 should be deleted

Conclusions:
- it seems there is correlation between age and sex - crabs of age of 1, 2, 3 months are more likely to be of the indeterminate sex than of the male one, while crabs over 24 months are much more likely to be male or female
- the kind of 'turning point' is the age of 5 months, after which most of the features increases for most of the instances
- there are yet some instances, where the age > 5 month doesn't mean the increase of the features, this pertains also to length and height - we can see that these features don't incraese so much with the age growth
- the shell weight + the shucked weight doesn't give in total the weight and supposedly they should - maybe the reason is the water, as observed in the abalone dataset (https://shapbio.me/courses/biolB215s14/abalone_cleaning.html); we will calculate the weight of the shell by substracting the shucked weight from the total weight and compare it with the weight of the shell; if any instance will have greater shell weight than the calculated shell weight, it will be deleted
