```python
import numpy as np

import numpy_financial as npf

import matplotlib.pyplot as plt

house_start_price = 120000
years = 5
month_numb = years*12
change = 0.05
change_month = change/12

invest_rate = 0.12
invest_rate_month = invest_rate/12

periods = np.arange(1,month_numb+1,dtype=int)

future_value = np.around(npf.fv(change_month, periods, 0, -house_start_price),2)

print("Wartość mieszkania po 5 latach to: {} zł.".format(future_value[-1]))

a = (((1+invest_rate_month)**month_numb)-1)

b = ((1+invest_rate_month)/invest_rate_month)

c = ((1+invest_rate_month)**month_numb)

d = b*a

e = d+c

monthly_payments = -future_value[-1]/-e

payments_cum = np.around(npf.fv(invest_rate_month,periods,-monthly_payments,0))

print("Wysokość miesięcznych wpłat wynosi {} zł.".format(round(monthly_payments,2)))

plt.plot(future_value,label='wartość mieszkania')
plt.plot(payments_cum,label='miesięczne wpłaty')
plt.legend()
plt.xlabel('Liczba okresów')
plt.ylabel('Wartość mieszkania')
```
