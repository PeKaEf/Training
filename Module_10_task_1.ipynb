```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data=pd.read_csv("C:/Users/pfali/Downloads/titanic (1).csv")
fare=data.loc[data['Fare'].notnull(),'Fare'].values
```

```python
fare_mean=np.round(np.mean(fare))
print(fare_mean)
```

```python
fare_median=np.round(np.median(fare))
print(fare_median)
```

```python
import scipy
from scipy import stats
fare_mode=stats.mode(fare)
print(fare_mode)
```

```python
q0=np.quantile(fare,0.0)
q1=np.quantile(fare,0.25)
q3=np.quantile(fare,0.75)
q4=np.quantile(fare,0.1)
print("q0 {}".format(q0))
print("q1 {}".format(q1))
print("q2 {}".format(fare_median))
print("q3 {}".format(q3))
print("q4 {}".format(q4))
```

```python
plt.figure(figsize=(15,7))
plt.hist(fare,bins=100)
plt.axvline(x=fare_mean,color='orange',label='Średnia')
plt.axvline(x=fare_median,color='red',label='Mediana')
plt.axvline(x=fare_mode[0],color='green',label='Moda')
plt.axvline(x=q0,color='black',label='Min')
plt.axvline(x=q1,color='black',label='Q1')
plt.axvline(x=q3,color='black',label='Q3')
plt.axvline(x=q4,color='black',label='Max')
plt.title('Rozkład cechy "Fare"')
plt.legend(loc='upper right')
plt.show()
```

```python
fare_range=max(fare)-min(fare)
print(fare_range)
```

```python
fare_iqr=q3-q1
print(fare_iqr)
```

```python
plt.boxplot(fare)
plt.show()
```

```python
fare_variance=np.var(fare,ddof=1)
print(fare_variance)
```

```python
fare_std=np.std(fare,ddof=1)
print(fare_std)
```

```python
import sklearn
from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()
fare_standarized=scaler.fit_transform(fare.reshape(-1,1))

plt.hist(fare_standarized,bins=100)
plt.axvline(x=fare_standarized.mean(),color='green',label='Średnia')
plt.show()
print("Średnia: {}".format(fare_standarized.mean()))
print("Odchylenie standardowe {}".format(np.std(fare_standarized,ddof=1)))
```
