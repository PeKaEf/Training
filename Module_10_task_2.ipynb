```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from numpy import random

a=random.normal(loc=3,scale=1,size=(25,4))
a
```

```python
b=random.normal(loc=3,scale=1,size=(100,100))
b
```

```python
a_mean=np.round(np.mean(a))
print(a_mean)
b_mean=np.round(np.mean(b))
print(b_mean)
```

```python
a_median=np.round(np.median(a))
print(a_median)
b_median=np.round(np.median(b))
print(b_median)
```

```python
import scipy
from scipy import stats
a_mode=stats.mode(a)
b_mode=stats.mode(b)
print(a_mode)
print(b_mode)
```

```python
aq0=np.quantile(a,0.0)
aq1=np.quantile(a,0.25)
aq3=np.quantile(a,0.75)
aq4=np.quantile(a,1)
print(f'A: Q0 {aq0}')
print(f'A: Q1 {aq1}')
print(f'A: Q2 {a_median}')
print(f'A: Q3 {aq3}')
print(f'A: Q4 {aq4}')
bq0=np.quantile(b,0.0)
bq1=np.quantile(b,0.25)
bq3=np.quantile(b,0.75)
bq4=np.quantile(b,1)
print(f'B: Q0 {bq0}')
print(f'B: Q1 {bq1}')
print(f'B: Q2 {b_median}')
print(f'B: Q3 {bq3}')
print(f'B: Q4 {bq4}')
```

```python
plt.figure(figsize=(10,6))
plt.hist(a,bins=10)
plt.axvline(x=a_mean,color='red',label='Średnia')
plt.axvline(x=a_median,color='green',label='Mediana')
plt.axvline(x=aq0,color='black',label='Min')
plt.axvline(x=aq1,color='black',label='Q1')
plt.axvline(x=aq3,color='black',label='Q3')
plt.axvline(x=aq4,color='black',label='Max')
plt.legend(loc='right',bbox_to_anchor=(1.18,0.85))
plt.title("Rozkład zbioru a")
plt.show()
```

```python
plt.figure(figsize=(20,6))
plt.hist(a,bins=10)
plt.axvline(x=b_mean,color='red',label='Średnia')
plt.axvline(x=b_median,color='green',label='Mediana')
plt.axvline(x=bq0,color='black',label='Min')
plt.axvline(x=bq1,color='black',label='Q1')
plt.axvline(x=bq3,color='black',label='Q3')
plt.axvline(x=bq4,color='black',label='Max')
plt.legend(loc='right',bbox_to_anchor=(1.18,0.85))
plt.title("Rozkład zbioru b")
plt.show()
```

```python
a_range=np.max(a)-np.min(a)
print(a_range)
b_range=np.max(b)-np.min(b)
print(b_range)
```

```python
a_iqr=aq3-aq1
print(a_iqr)
b_iqr=bq3-bq1
print(b_iqr)
```

```python
a_variance=np.var(a,ddof=1)
print(a_variance)
b_variance=np.var(b,ddof=1)
print(b_variance)
```

```python
a_std=np.std(a,ddof=1)
print(a_std)
b_std=np.std(b,ddof=1)
print(b_std)
```
