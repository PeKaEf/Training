```python
data['DateofHire']=pd.to_datetime(data['DateofHire'])
data['DateofTermination']=pd.to_datetime(data['DateofTermination'])
data['current_date']="2023-09-07"
data['current_date']=pd.to_datetime(data['current_date'])
```

```python
data.dtypes
```

```python
data['Seniority']=data['current_date']-data['DateofHire']
pd.set_option('display.max_rows',None)
data['Seniority']
```

```python
data['Seniority_days']=data['Seniority'].dt.days
data['Seniority_years']=data['Seniority_days']/365
data['Seniority_years']
```

```python
seniority=data[['Seniority_years','RecruitmentSource']].groupby(['RecruitmentSource']).mean().reset_index()
seniority
```

```python
plt.figure(figsize=(20,8))
plt.title('StaÅ¼ pracy a ÅºrÃ³dÅ‚o pozyskania pracownika')
g=sns.barplot(data=seniority,x='RecruitmentSource',y='Seniority_years',color='orange')
plt.ylabel("Åšredni staÅ¼ pracownika w latach")
plt.xlabel("Å¹rÃ³dÅ‚o pozyskania pracownika")
g.set_xticklabels(g.get_xticklabels(),rotation=90)
plt.show
```

```python
plt.figure(figsize=(15,10))
plt.title('Zadowolenie z pracy a stan cywilny pracownika')
p=sns.countplot(x=data['EmpSatisfaction'],hue=data['MaritalDesc'])
plt.xlabel("Ocena zadowolenia pracownika")
plt.ylabel("Liczba pracownikÃ³w")
p.legend(['zamÄ™Å¼na/Å¼onaty','rozwiedziony','singiel','owdowiaÅ‚y','w separacji'])
plt.show
```

```python

data['DOB']=pd.to_datetime(data['DOB'],format='mixed')
data['Age']=(data['current_date']-data['DOB'])
data['Age_days']=data['Age'].dt.days
data['Age_years']=data['Age_days']/365
data['Age_years'].astype('int64')
```

```python
df=data.loc[data["DateofTermination"].isna(),"Age_years"]
df_age=pd.DataFrame(df)
df_age['Occurrences']=1
df_age['Age_years']=np.floor(pd.to_numeric(df_age['Age_years'],errors='coerce')).astype("Int64")
df_age.dtypes
```

```python
plt.figure(figsize=(20,12))
plt.title('Struktura wiekowa aktualnie zatrudnionych pracownikÃ³w')
sns.countplot(data=df_age,x='Age_years',color='lightskyblue')
plt.xlabel("Wiek pracownikÃ³w w latach")
plt.ylabel("Liczba pracownikÃ³w")
```
