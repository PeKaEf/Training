We have 83 features of various data type. Null values need to be completed. At first glance almost all features can contribute to the target, thus, initially we won't
exclude any feature.

Completing null values

MsZoning - this is nominal data, with 5 values and only 4 missing, which will be imputed with mode

LotFrontage - this is numerical, ratio, continuous data

Alley - nominal data, with two values: Pave, Grvl; 1369 out of 1460 values missing? - might be not missing but 'not applicable', let's try to  leave the feature first with transforming it to categorical data type

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn import preprocessing

train_df['MSZoning'].fillna(train_df['MSZoning'].mode()[0],inplace=True)

imp = IterativeImputer(max_iter=10, random_state = 0)

train_df['LotFrontage'] = imp.fit_transform(np.array(train_df['LotFrontage']).reshape(-1,1))
test_df['LotFrontage'] = imp.fit_transform(np.array(test_df['LotFrontage']).reshape(-1,1))

train_df['Alley'].fillna("Not",inplace=True)
test_df['Alley'].fillna("Not",inplace=True)

le = preprocessing.LabelEncoder()
train_df['Alley'] = le.fit_transform(train_df['Alley'])
test_df['Alley'] = le.fit_transform(test_df['Alley'])

train_df.head()
```

Exterior1st and Exterior2nd - 1 missing value per each in the test dataset, indicating 'not applicable' feature; will be replaced with missing and then transformed to category.

```python
test_df['Exterior1st'].fillna("Missing",inplace=True)
test_df['Exterior2nd'].fillna("Missing",inplace=True)
```

MasVnrType and MasVnrArea - None in MasVnrType in the training dataset means absence, which is indicated by 0 in MasVnrArea; therefore, we have in fact only 8 missing values in both features, thus, they will be imputed with mode

```python
train_df['MasVnrType'].fillna(train_df['MasVnrType'].mode()[0],inplace=True)
train_df['MasVnrArea'].fillna(train_df['MasVnrArea'].mode()[0],inplace=True)
```

MasVnrType and MasVnrArea - in the test dataset 15 null values of MasVnrArea indicate 'not applicable', so to be replaced with zeros. However, MasVnrType have more missing values, thus, possibly it will be better to use KNNImputer to impute nulls.

```python
test_df['MasVnrArea'].fillna(0,inplace=True)

from sklearn import preprocessing
from sklearn.impute import KNNImputer

imputer = KNNImputer(n_neighbors=1460)
le = preprocessing.LabelEncoder()

test_df['MasVnrType'] = le.fit_transform(test_df['MasVnrType'])
test_df['MasVnrType'] = imputer.fit_transform(test_df[['MasVnrType']]).ravel()
```

BsmtQual, BsmtCond, BsmtFinType1 have 37 missing values each which equals to the number of 0-value data in TotalBsmtSF, which means missing values indicate no basement. Thus, can be filled with "Missing". In the test dataset it is similar - 42 zeros in TotalBsmtSF against 44, 45, 44, 42 and 42 nulls in BsmtQual, BsmtCond, BsmtFinType1, BsmtExposure, BsmtFinType2, so all will be imputed with "Missing" and then transformed to categories.

```python
train_df['BsmtQual'].fillna("Missing",inplace=True)
train_df['BsmtCond'].fillna("Missing",inplace=True)
train_df['BsmtFinType1'].fillna("Missing",inplace=True)

test_df['BsmtQual'].fillna("Missing",inplace=True)
test_df['BsmtCond'].fillna("Missing",inplace=True)
test_df['BsmtFinType1'].fillna("Missing",inplace=True)
test_df['BsmtExposure'].fillna("Missing",inplace=True)
test_df['BsmtFinType2'].fillna("Missing",inplace=True)
```

BsmtExposure, BsmtFinType2 - both have 1 null value in the training dataset, but seems to be dependent of other basement-describing features, thus, to be replaced using KNNImputer
BsmtUnfSF, TotalBsmtSF - both have 1 null value in the test dataset, but seems to be dependent of other basement-describing features, thus, to be replaced using KNNImputer

```python
imputer = KNNImputer(n_neighbors=1460)
le = preprocessing.LabelEncoder()

train_df['BsmtExposure'] = le.fit_transform(train_df['BsmtExposure'])
train_df['BsmtFinType2'] = le.fit_transform(train_df['BsmtFinType2'])

train_df['BsmtExposure'] = imputer.fit_transform(train_df[['BsmtExposure']]).ravel()
train_df['BsmtFinType2'] = imputer.fit_transform(train_df[['BsmtFinType2']]).ravel()

test_df['BsmtUnfSF'] = le.fit_transform(test_df['BsmtUnfSF'])
test_df['TotalBsmtSF'] = le.fit_transform(test_df['TotalBsmtSF'])

test_df['BsmtUnfSF'] = imputer.fit_transform(test_df[['BsmtUnfSF']]).ravel()
test_df['TotalBsmtSF'] = imputer.fit_transform(test_df[['TotalBsmtSF']]).ravel()
```

Electrical - this is nominal data with 1 null value; to be imputed with mode
BsmtFullBath, BsmtHalfBath, Functional - these are ratio, continuous data with 1 or 2 null values; to be imputed with mode

```python
train_df['Electrical'].fillna(train_df['Electrical'].mode()[0],inplace=True)

test_df['BsmtFullBath'].fillna(test_df['BsmtFullBath'].mode()[0],inplace=True)
test_df['BsmtHalfBath'].fillna(test_df['BsmtHalfBath'].mode()[0],inplace=True)
test_df['Functional'].fillna(test_df['Functional'].mode()[0],inplace=True)
```

Missing values for FireplaceQu mean in fact 'not applicable' feature (=no fireplace), thus, can be replaced with 0. The same for features related to garages: GarageType, GarageYrBlt, GarageFinish, GarageQual, GarageCond. These applies to both the test and training dataset. The same for PoolQC - nulls indicate no pool, but only in the training dataset.

```python
train_df['FireplaceQu'].fillna("Missing",inplace=True)
train_df['GarageType'].fillna("Missing",inplace=True)
train_df['GarageYrBlt'].fillna(0,inplace=True)
train_df['GarageFinish'].fillna("Missing",inplace=True)
train_df['GarageQual'].fillna("Missing",inplace=True)
train_df['GarageCond'].fillna("Missing",inplace=True)
train_df['PoolQC'].fillna("Missing",inplace=True)

test_df['FireplaceQu'].fillna("Missing",inplace=True)
test_df['GarageType'].fillna("Missing",inplace=True)
test_df['GarageYrBlt'].fillna(0,inplace=True)
test_df['GarageFinish'].fillna("Missing",inplace=True)
test_df['GarageQual'].fillna("Missing",inplace=True)
test_df['GarageCond'].fillna("Missing",inplace=True)
```

Fence and MiscFeature - both have many missing values, which can be either nulls or 'not applicable' values. Therefore, both will be imputed with KNNImputer. Additionally, PoolQC will be imputed the same way in the test dataset.
The same way GarageCars and GarageArea in the test dataset will be imputed, because they can be correlated with other car-type features.

```python
train_df['Fence'] = le.fit_transform(train_df['Fence'])
train_df['MiscFeature'] = le.fit_transform(train_df['MiscFeature'])
train_df['Fence'] = imputer.fit_transform(train_df[['Fence']]).ravel()
train_df['MiscFeature'] = imputer.fit_transform(train_df[['MiscFeature']]).ravel()

test_df['Fence'] = le.fit_transform(test_df['Fence'])
test_df['MiscFeature'] = le.fit_transform(test_df['MiscFeature'])
test_df['Fence'] = imputer.fit_transform(test_df[['Fence']]).ravel()
test_df['MiscFeature'] = imputer.fit_transform(test_df[['MiscFeature']]).ravel()
test_df['PoolQC'] = le.fit_transform(test_df['PoolQC'])
test_df['PoolQC'] = imputer.fit_transform(test_df[['PoolQC']]).ravel()

test_df['GarageCars'] = le.fit_transform(test_df['GarageCars'])
test_df['GarageArea'] = le.fit_transform(test_df['GarageArea'])
test_df['GarageCars'] = imputer.fit_transform(test_df[['GarageCars']]).ravel()
test_df['GarageArea'] = imputer.fit_transform(test_df[['GarageArea']]).ravel()
```

SaleType - this is nominal data with 1 value missing in the test dataset, thus, to be replaced with mode.

```python
test_df['SaleType'].fillna(test_df['SaleType'].mode()[0],inplace=True)
```
